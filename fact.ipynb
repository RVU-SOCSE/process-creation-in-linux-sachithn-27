```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
num = int(input("Enter a number"))
if num<0:
  print("Factorial not defined")
else:
  print("Factorial is :", factorial(num))
```

```python
def fib(n):
    if n<=1:
        return n
    else:
        return fib(n-1) + fib(n-2)
print(fib(8))
```
